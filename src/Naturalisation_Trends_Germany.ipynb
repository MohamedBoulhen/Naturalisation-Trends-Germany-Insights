# <span style="color:darkblue">2. Data Loading and Preprocessing

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
I start by loading the dataset from  containing information about naturalisations in Germany. The dataset includes details such as the year of naturalisation, country of origin, gender, and age group of naturalised citizens. Before conducting our analysis, i perform preprocessing steps to clean and prepare the dataset for analysis. This involves handling missing values, formatting data types, and ensuring data consistency.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
path = '/Users/mo/Desktop/RSE/citizenshipger_trends-main/data/test1.csv'
df = pd.read_csv(path, encoding = 'latin1', engine = 'python', delimiter = ';')

# Display the first few rows of the dataset
#print(df.head())
print(df.columns)
```

# <span style="color:darkblue">3. Analysis

## &emsp;<span style="color:darkblue">First Question: Evolution of Naturalisation Numbers Over the Years

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
I begin by examining the trend in naturalisation numbers from 2010 to 2022. This allows us to understand how the total number of naturalisations has changed over time, providing insights into immigration patterns and integration efforts in Germany.
</p>


```python
# Filter rows with "Total" in the "citizenship" column
total_df = df[df["citizenship"] == "Total"]

# Group by year and calculate the sum of "Gesamt"
sum_gesamt_by_year = total_df.groupby("years")["Gesamt"].sum()

# Display the result
print(sum_gesamt_by_year)
```

```python
# Plot the total number of naturalisations by year
plt.figure(figsize=(10, 6))
plt.plot(sum_gesamt_by_year.index, sum_gesamt_by_year.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Number of Naturalisations')
plt.title('Total Number of Naturalisations in Germany (2010 - 2022)')
plt.grid(True)
plt.show()
```

### Conclusion - Evolution of Naturalisation Numbers Over the Years:

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
The trend analysis of naturalisation numbers from 2010 to 2022 reveals several noteworthy patterns. Initially, from 2010 to 2014, we observe relatively stable figures, indicating a consistent but not rapid pace of naturalisations. However, a slight decline is noticed in 2015, followed by a subsequent rebound in 2016.
A more substantial increase is evident from 2017 onwards, with a notable surge in 2019 and 2022. Particularly, 2019 and 2022 stand out as peak years, with 124,165 and 159,505 naturalisations, respectively. This surge suggests a potential shift in immigration patterns or heightened integration efforts during these periods.

## &emsp;<span style="color:darkblue"> Second Question: Origins/Regions of the Most Naturalised Citizens in Germany

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Let's explore the origins or regions that contribute the most to naturalised citizenship in Germany. We'll focus on continents or major regions such as Africa, America, Asia, Australia and Oceania, and Europe. Analyzing the distribution of naturalised citizens by origin provides insights into the diversity of immigrants and their integration into German society over the years.

```python
# Filter out rows where citizenship is not equal to 'Total'
df_filtered = df[df['citizenship'] != 'Total']

# Filter only the specified citizenships/origin
specified_citizenships = ['Africa', 'America', 'Asia', 'Australia and Oceania', 'Europe']
df_specified = df_filtered[df_filtered['citizenship'].isin(specified_citizenships)]

# Group data by citizenship and calculate the sum of 'Gesamt' for all years
grouped_data = df_specified.groupby('citizenship')['Gesamt'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['citizenship'], grouped_data['Gesamt'], color='skyblue')
plt.xlabel('Origin')
plt.ylabel('Total Count')
plt.title('Total Counts of Citizens 2010 - 2022')
plt.xticks(rotation=45)
plt.show()
```

### Conclusion - Origins of the Most Naturalised Citizens in Germany:

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
In summary, Europe dominates the naturalization process in Germany, followed by Asia, America, Africa, and a negligible presence from Australia and Oceania.

## &emsp;<span style="color:darkblue"> Extra Question: About Origins/Regions of the Most Naturalised Citizens in Germany

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Moving on to explore citizens from origins other than Africa, America, Asia, Australia and Oceania, and Europe, we filter out rows corresponding to these regions. This allows us to focus on understanding the naturalisation trends of individuals from diverse and less represented origins. By grouping the data based on citizenship and calculating the total count of naturalised citizens for each origin, we gain insights into the contribution of these lesser-known regions to the overall naturalisation process in Germany over the years.


```python
# Exclude specified origin
excluded_citizenships = ['Africa', 'America', 'Asia', 'Australia and Oceania', 'Europe']
df_excluded = df_filtered[~df_filtered['citizenship'].isin(excluded_citizenships)]

# Group data by citizenship and calculate the sum of 'Gesamt'
grouped_data = df_excluded.groupby('citizenship')['Gesamt'].sum()

# Plot the data
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', color='dodgerblue')

plt.xlabel('Region')
plt.ylabel('Total Count')
plt.title('Total Counts of Citizens 2010 - 2022')
plt.xticks(rotation=45, ha='right')  # Rotate the x-labels for better readability
plt.tight_layout()
plt.show()
```

### Conclusion - Origins/Regions of the Most Naturalised Citizens in Germany:

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
The region of South and South East Asia contributes the most to naturalized citizenship in Germany. With a total count exceeding 200,000, it stands out as the primary source of immigrants becoming German citizens.
Factors such as economic opportunities, education, and family ties likely drive this trend maybe.

## &emsp;<span style="color:darkblue"> Third Question: Gender Differences Among Naturalisers


<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Exploring potential gender differences among naturalisers to assess whether there are disparities in the naturalisation process based on gender. Understanding any gender imbalances in naturalisation rates is essential for promoting equality and inclusivity in the citizenship process.

```python
# Calculate total counts of males and females
total_male = total_df['Male Gesamt'].sum()
total_female = total_df['Female Gesamt'].sum()

# Calculate percentages
percentage_male = (total_male / (total_male + total_female)) * 100
percentage_female = (total_female / (total_male + total_female)) * 100

# Plot the percentages in a pie chart
labels = ['Male', 'Female']
sizes = [percentage_male, percentage_female]
colors = ['dodgerblue', 'pink']
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Gender Composition Among Naturalisers (Overall 2010 - 2022)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Let's check the difference between the percentages of females and males among naturalisers for each year. This will allow us to identify any variations or trends in gender representation over time.

```python
gender_totals = df_filtered.groupby('years')[['Male Gesamt', 'Female Gesamt']].sum()

# Convert the index to strings
years = gender_totals.index.astype(str)

# Set up the figure and axes
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = np.arange(len(years))

# Plot bars for male counts
plt.bar(index, gender_totals['Male Gesamt'], bar_width, label='Male', color='dodgerblue', alpha=0.7)

# Plot bars for female counts
plt.bar(index + bar_width, gender_totals['Female Gesamt'], bar_width, label='Female', color='pink', alpha=0.7)

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Total Number of Naturalizations')
plt.title('Gender Distribution of Naturalized Citizens by Year from 2010 to 2022')
plt.xticks(index + bar_width / 2, years)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

### Conclusion - Gender Differences Among Naturalisers:
<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Gender Parity (2010-2021):
From 2010 to 2021, the naturalization rates for males and females were nearly equal. Both genders contributed similarly to the overall count of naturalized citizens.
This gender parity suggests that the naturalization process was relatively equitable during this period.
<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Spike in Male Naturalizations (2022):
In 2022, there was a significant spike in male naturalizations. The blue bar representing males is notably higher than the pink bar representing females.

## &emsp;<span style="color:darkblue"> Fourth Question: Distribution of Naturalised Citizens Across Age Groups

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Analysing the distribution of naturalised citizens across different age groups provides insights into integration patterns, workforce dynamics, and the societal contributions of immigrants to Germany.

```python
# Define age columns representing the total number of naturalized citizens for each age group
age_columns = ['Total under 5 years', 'Total 5 to under 10 years', 'Total 10 to under 15 years',
               'Total 25 to under 30 years', 'Total 30 to under 35 years', 'Total 35 to under 40 years',
               'Total 40 to under 45 years', 'Total 45 to under 50 years', 'Total 50 to under 55 years',
               'Total 55 to under 60 years']

# Make a copy of the filtered DataFrame
df_filtered_copy = df_filtered.copy()

# Convert the selected columns to numeric data type
df_filtered_copy[age_columns] = df_filtered_copy[age_columns].apply(pd.to_numeric, errors='coerce')

# Sum the total number of naturalized citizens for each age group
total_naturalized_by_age = df_filtered_copy[age_columns].sum()

# Plot the total number of naturalized citizens by age group
plt.figure(figsize=(12, 8))

plt.bar(age_columns, total_naturalized_by_age, color='skyblue')

plt.title('Total Naturalized Citizens by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Naturalized Citizens')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

-> The dominant age group for naturalized citizens in Germany lies between 30 - 35 years,

### &emsp;<span style="color:darkblue"> Fifth Question: Distribution of Naturalised Citizens Across Gender and Age Group

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
In this section, i would like to explore the distribution of naturalized citizens across different age groups and genders. I will analyze the total number of naturalized citizens for each age group and gender category to gain insights into demographic patterns among individuals who have undergone the naturalization process. By visualizing this data, we can better understand how naturalization varies across age demographics and gender identities, providing valuable insights into integration dynamics and societal contributions.

```python
# Define age and gender columns
age_columns = ['under 5 years', '5 to under 10 years', '10 to under 15 years',
              '25 to under 30 years', '30 to under 35 years', '35 to under 40 years',
              '40 to under 45 years', '45 to under 50 years', '50 to under 55 years',
              '55 to under 60 years', '60 to under 65 years', '65 to under 70 years',
              '70 to under 75 years']

# Define male and female columns
male_columns = [f'Male {age}' for age in age_columns]
female_columns = [f'Female {age}' for age in age_columns]

# Convert the selected columns to numeric data type
df_filtered_copy[male_columns + female_columns] = df_filtered_copy[male_columns + female_columns].apply(pd.to_numeric, errors='coerce')

# Sum the total number of naturalized citizens for each age group and gender
total_naturalized_by_age_gender = df_filtered_copy[male_columns + female_columns].sum()

# Plot the total number of naturalized citizens by age group and gender
plt.figure(figsize=(12, 8))
bar_width = 0.35
index = range(len(age_columns))

plt.bar(index, total_naturalized_by_age_gender[:len(age_columns)], bar_width, color='dodgerblue', label='Male')
plt.bar([i + bar_width for i in index], total_naturalized_by_age_gender[len(age_columns):], bar_width, color='pink', label='Female')

plt.title('Total Naturalized Citizens by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Total Naturalized Citizens')
plt.xticks([i + bar_width/2 for i in index], age_columns, rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```


### Conclusion - Gender Differences Among Naturalisers:
<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
t's evident that within the age group 30 to 35, males significantly outnumber females in terms of naturalization. Conversely, in the age range of 40 to 45, the trend reverses, with females surpassing males in naturalization rates. This indicates a fluctuating gender dynamic in naturalization patterns across different age brackets.

### &emsp;<span style="color:darkblue">Sixth Question: Analyzing Citizen Distribution Across German States

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
So which German states have the highest number of naturalised citizens?
I'll focus on regional differences. I'll identify the German states with the largest populations of naturalised citizens, highlighting areas where integration and diversity play significant roles.

```python
# Group the data by states and sum the total counts
total_counts_by_state = df_filtered.groupby('states')['Gesamt'].sum()

# Define custom colors for the bars, making this little colorful :)
custom_colors = ['salmon', 'peachpuff', 'gold', 'lightsalmon', 'tomato', 'orangered', 'hotpink', 'lightpink']

# Plot the total counts for each state with custom colors
plt.figure(figsize=(12, 8))
total_counts_by_state.sort_values(ascending=False).plot(kind='bar', color=custom_colors)
plt.title('Total Naturalized Citizens by State (Overall 2010 - 2022)')
plt.xlabel('State')
plt.ylabel('Total Citizens')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
This looks interesting, Nordrhein-Westfalen stands out with over 500,000 naturalized citizens between 2010 and 2022.

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Let's analyze the gender distribution within this region. By examining the number of male and female naturalized citizens separately, we can get insights into the demographic composition and integration patterns within Nordrhein-Westfalen. Let's proceed with our analysis to uncover the gender dynamics among naturalized citizens in this region.

```python
# Filter the data for Nordrhein-Westfalen
nordhein_westfalen_data = df_filtered[df_filtered['states'] == 'Nordrhein-Westfalen']

# Calculate the total number of naturalised citizens in Nordrhein-Westfalen
total_naturalized_citizens = nordhein_westfalen_data['Gesamt'].sum()

# Calculate the number of male naturalised citizens
total_male_citizens = nordhein_westfalen_data['Male Gesamt'].sum()

# Calculate the number of female naturalised citizens
total_female_citizens = nordhein_westfalen_data['Female Gesamt'].sum()

# Present the results
print("Total Number of Naturalised Citizens in Nordrhein-Westfalen:", total_naturalized_citizens)
print("Number of Male Naturalised Citizens in Nordrhein-Westfalen:", total_male_citizens)
print("Number of Female Naturalised Citizens in Nordrhein-Westfalen:", total_female_citizens)
```

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Let's create a bar plot showing the total number of naturalized citizens in Nordrhein-Westfalen, as well as the number of male and female naturalized citizens separately.

```python
# Data for plotting
categories = ['Total', 'Male', 'Female']
counts = [total_naturalized_citizens, total_male_citizens, total_female_citizens]

# Plot
plt.figure(figsize=(8, 6))
plt.bar(categories, counts, color=['skyblue', 'blue', 'pink'])
plt.title('Naturalised Citizens in Nordrhein-Westfalen')
plt.xlabel('Category')
plt.ylabel('Number of Citizens')
plt.show()
```

<p style="text-align: justify; font-size: 17px; line-height: 1.6;">
Better to create a pie chart showing the percentages of male and female naturalized citizens in Nordrhein-Westfalen.

```python
# Data for plotting
new_categories = ['Male', 'Female']
new_counts = [total_male_citizens, total_female_citizens]

# Calculate total count
total = sum(new_counts)

# Calculate percentages
percentages = [count / total * 100 for count in new_counts]

# Plot
plt.figure(figsize=(8, 6))
plt.pie(percentages, labels=new_categories, autopct='%1.1f%%', colors=['skyblue', 'pink'])
plt.title('Gender Distribution of Naturalised Citizens in Nordrhein-Westfalen')
plt.show()
```
